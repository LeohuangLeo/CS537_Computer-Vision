```python
from docx import Document
import matplotlib.pyplot as plt
```

```python
document = Document('history_2000_1.docx')
```

```python
acc = []
for i in range(1000000000000000):
    try:
        idx = document.paragraphs[i].text.find('Loss:')
        if idx != -1:
            acc.append(float(document.paragraphs[i].text[idx+6:idx+18]))
    except:
        break
```

```python
len(acc)
```

```python
plt.plot(range(1, len(acc)+1), acc)
plt.title('Model_best_bz12_lr_10e-5_ep_2001.pth')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show
```
